```python
from pylab import imshow, show, get_cmap
import matplotlib.pyplot as plt
from PIL import Image
import random
background = Image.open('images/black_background.png')
red = Image.open('coin82.png')

i=1
while i<=50:
    positionxred = random.randint(10, i+100) 
    positionyred = random.randint(10, i+150)
    background.paste(red,(positionxred+i, positionyred+i))
    i+=10

background.save("abcd1.png")
background.show("abcd1.png")
```

```python
# read the image in 
photo=plt.imread("abcd1.png")
#lET US SEE THE SHAPE OF THE IMAGE.
print(photo.shape)
plt.imshow(photo)
```
